## 1. Affichage en fonction de la méthode de longitude utilisée
---

Chargement des librairies

```julia
using PlotlyJS
using DelimitedFiles
using Pkg
using LaTeXStrings
```

Récupération des données 

```julia
### Récupération des données ###
IRR_data = readdlm("ellipt20.res") #importing data 

year = IRR_data[5:605 , 1]
IRR_trl = IRR_data[5:605, 2]
IRR_cal = IRR_data[610:1210 ,2]
```

Méthode de la longitude exacte

```julia
### Figures ###
p1 = plot(-600:1:0, IRR_trl, Layout(xaxis_title = "Year BP", yaxis_title ="Irradiance [MJ/m²]", 
                    title = attr(text ="Solar Irradiance between true longitude 0° and 90° at 65°N for the last 600 ka", y=0.98,x=0.5,xanchor= "center",yanchor= "top"), font =attr(size = 20)))
```

Méthode de la longitude moyenne

```julia
### Figures ###
p2 = plot(-600:1:0, IRR_cal, line = attr(color="firebrick"), Layout(xaxis_title = "Year BP", yaxis_title ="Irradiance [MJ/m²]", 
                    title = attr(text ="Solar Irradiance between March 21 and June 22 at 65°N for the last 600 ka", y=0.98,x=0.5,xanchor= "center",yanchor= "top"), font =attr(size = 20)))
```

---
## 2. Comparaison entre les deux méthodes
---

```julia
### Figures ###
IRR_diff = IRR_trl - IRR_cal
p1 = plot(-600:1:0, IRR_diff, line = attr(color="green"), Layout(xaxis_title = "Year BP", yaxis_title ="Irradiance [MJ/m²]", 
                    title = attr(text ="Difference in Solar Irradiance between true longitude method and mean longitude method at 65°N for the last 600 ka", y=0.98,x=0.5,xanchor= "center",yanchor= "top"), font =attr(size = 16)))
```
